Question: What proportion of Netflix content is movies vs series?
conclusion: netflix usally has more movies than tv shows

```python
sns.countplot(x='type', data=df)
plt.title('Distribution of Movie and TV Show')
```

Question: How has Netflix content grown each year?
conclusion:Netflix expansion strategy and growth period.

```python
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("Content Added to Netflix Over Time")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles Added")
```

Question: Which countries contribute most content?
conclusion:United states release highest no of content on netflix

```python
top_countries = df['country'].value_counts().head(3)
top_countries.plot(kind='bar')
plt.title("Top 3 Countries with Most Content on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
```

Question: What type of audience Netflix targets?
conclusion: Tv_MA

```python
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title("Distribution of Netflix Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

Question: What genres Netflix offers most?
conclusion:International movies

```python
genres = df['listed_in'].str.split(', ').explode()

genres.value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```
